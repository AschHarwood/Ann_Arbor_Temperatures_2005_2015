The GHCN-Daily is comprised of daily climate records from thousands of land surface stations across the globe.

Each row in the assignment datafile corresponds to a single observation.

The following variables are provided to you:

* **id** : station identification code
* **date** : date in YYYY-MM-DD format (e.g. 2012-01-24 = January 24, 2012)
* **element** : indicator of element type
    * TMAX : Maximum temperature (tenths of degrees C)

* **value** : data value for element (tenths of degrees C)

```python
import matplotlib.pyplot as plt
import mplleaflet
import pandas as pd

def leaflet_plot_stations(binsize, hashid):

    df = pd.read_csv('data/C2A2_data/BinSize_d{}.csv'.format(binsize))

    station_locations_by_hash = df[df['hash'] == hashid]

    lons = station_locations_by_hash['LONGITUDE'].tolist()
    lats = station_locations_by_hash['LATITUDE'].tolist()

    plt.figure(figsize=(8,8))

    plt.scatter(lons, lats, c='r', alpha=0.7, s=200)

    return mplleaflet.display()

leaflet_plot_stations(400,'fb441e62df2d58994928907a91895ec62c2c42e6cd075c2700843b89')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#reading in the dataframe
```

```python
df = pd.read_csv('fb441e62df2d58994928907a91895ec62c2c42e6cd075c2700843b89.csv')
```

```python
#converting date to datetime object
df['Date'] = pd.to_datetime(df['Date'])
```

```python
#function for finding the leap years
def get_leap_year(date):
    if date['day']==29 and date['month']==2:
        return True
    else:
        return False
```

```python
#grabs day
df['day'] = df['Date'].dt.day
```

```python
#grabs month
df['month'] = df['Date'].dt.month
```

```python
#removes leap years
df['leap_year']=df.apply(lambda x: get_leap_year(x), axis=1)
```

```python
#removes leap years
df_no_leap = df[df['leap_year']==False]
```

```python
#gets temp max values
max_temp = df_no_leap[df_no_leap['Element']=='TMAX']
```

```python
#gets temp min values
min_temp = df_no_leap[df_no_leap['Element']=='TMIN']
```

```python
#averages the temp for each day
max_temp_avg = max_temp.groupby('Date').mean()
```

```python
#Averages min temp for each day
min_temp_avg = min_temp.groupby('Date').mean()
```

```python
#removes 2015 dates
max_temp_avg = max_temp_avg.reset_index()
max_temp_avg = max_temp_avg[max_temp_avg['Date'].dt.year != 2015]
```

```python
#finds the max temp for each day of the year
max_temp_df = max_temp_avg.groupby(['month', 'day']).max()
```

```python
#finds min temp
min_temp_avg = min_temp_avg.reset_index()
min_temp_avg = min_temp_avg[min_temp_avg['Date'].dt.year != 2015]

min_temp_df = min_temp_avg.groupby(['month', 'day']).min()
```

```python
#converts dates to teh new string format
min_temp_df['month_day'] = min_temp_df['Date'].apply(lambda x: x.strftime('%m-%d'))
min_temp_df['month_day'] = pd.to_datetime(min_temp_df['month_day'], format=('%m-%d'))
```

```python
max_temp_df['month_day'] = max_temp_df['Date'].apply(lambda x: x.strftime('%m-%d'))
max_temp_df['month_day'] = pd.to_datetime(max_temp_df['month_day'], format=('%m-%d'))
```

```python
#creates a new date range for matplotlip fill_between
d = max_temp_df['month_day'].values
```

```python
#generates my fig
fig = plt.figure(figsize = (10,10))
plt.plot(max_temp_df['month_day'], max_temp_df['Data_Value'])
plt.plot(min_temp_df['month_day'], min_temp_df['Data_Value'])
plt.fill_between(d, max_temp_df['Data_Value'],min_temp_df['Data_Value'], alpha=0.25)
```

```python
#removes leap year for 2015 df
max_temp_2015 = df_no_leap[df_no_leap['Element']=='TMAX']
min_temp_2015 = df_no_leap[df_no_leap['Element']=='TMIN']
```

```python
#repeats cleaning from above but for 2015
max_temp_avg_2015 = max_temp_2015.groupby('Date').mean()
min_temp_avg_2015 = min_temp_2015.groupby('Date').mean()
```

```python
max_temp_avg_2015.head()
```

```python
max_temp_avg_2015 = max_temp_avg_2015.reset_index()
max_temp_avg_2015 = max_temp_avg_2015[max_temp_avg_2015['Date'].dt.year == 2015]

max_temp_df_2015 = max_temp_avg_2015.groupby(['month', 'day']).max()

min_temp_avg_2015 = min_temp_avg_2015.reset_index()
min_temp_avg_2015 = min_temp_avg_2015[min_temp_avg_2015['Date'].dt.year == 2015]

min_temp_df_2015 = min_temp_avg_2015.groupby(['month', 'day']).min()
```

```python
min_temp_df_2015['month_day'] = min_temp_df_2015['Date'].apply(lambda x: x.strftime('%m-%d'))
min_temp_df_2015['month_day'] = pd.to_datetime(min_temp_df_2015['month_day'], format=('%m-%d'))
max_temp_df_2015['month_day'] = max_temp_df_2015['Date'].apply(lambda x: x.strftime('%m-%d'))
max_temp_df_2015['month_day'] = pd.to_datetime(max_temp_df_2015['month_day'], format=('%m-%d'))
```

```python
max_temp_df_2015.rename(columns={'Data_Value': 'Max_temp_2015', "month_day": "2015_month_day"}, inplace=True)
```

```python
max_join = pd.concat([max_temp_df, max_temp_df_2015], axis=1)
```

```python
max_join['2015_larger'] = max_join['Data_Value'] < max_join['Max_temp_2015']
```

```python
Temp_max_outliers_2015 = max_join[max_join['2015_larger']==True]
```

```python
min_temp_df_2015.rename(columns={'Max_temp_2015': 'Min_temp_2015','month_day': '2015_month_day'}, inplace=True)
```

```python
min_join = pd.concat([min_temp_df, min_temp_df_2015], axis=1)
```

```python
min_join['2015_smaller'] = min_join['Data_Value'] > min_join['Min_temp_2015']
```

```python
Temp_min_outliers_2015 = min_join[min_join['2015_smaller']==True]
```

```python
from matplotlib.dates import MonthLocator, DateFormatter
```

```python
import matplotlib.dates as mdates
```

```python
import datetime
```

```python
fig,ax = plt.subplots(figsize=(15,5))
ax.plot(max_temp_df['month_day'], max_temp_df['Data_Value']/10, c='#E3655B', label='High Temperatures 2005-2014')
ax.plot(min_temp_df['month_day'], min_temp_df['Data_Value']/10, c='#5FA5EA', label='Low Temperatures 2005-2014')
ax.fill_between(d, max_temp_df['Data_Value']/10,min_temp_df['Data_Value']/10, alpha=0.20, facecolor='#EEE5E9')
ax.scatter(Temp_max_outliers_2015['2015_month_day'].values, Temp_max_outliers_2015['Max_temp_2015']/10, c='#38618C', label='High Temperatures 2015')
ax.scatter(Temp_min_outliers_2015['2015_month_day'].values, Temp_min_outliers_2015['Min_temp_2015']/10, c='#38618C', label='Low Temperatures 2015')

ax.xaxis.set_major_locator(mdates.MonthLocator())

monthFmt = mdates.DateFormatter('%b')
ax.xaxis.set_major_formatter(monthFmt)
ax.set_xlim([datetime.date(1900, 1, 1), datetime.date(1900, 12, 31)])
ax.set_ylabel('Celsius', alpha=.7)
ax.set_frame_on(False)
#plt.title('Record High and Low Temperatures in Ann Arbor, Michigan 2005 - 2015')
ax.set_title('Record High and Low Temperatures in Ann Arbor, Michigan 2005 - 2015', fontsize=14, alpha=.7)
leg = plt.legend(loc='best',frameon=False)
for text in leg.texts:
    text.set_alpha(.7)
plt.xticks(alpha=.7)
plt.yticks(alpha=.7)

plt.savefig('temps_ann_arbor.png')


```
